```python
## data-1
df1 = pd.DataFrame(np.random.randn(100,4), columns= 'A B C D'.split())
df1.head()
```

```python
#Data-2
df2= pd.DataFrame({
    'Category':['A','B','C'],
    'values':[32,43,50]
})
df2
```

```python
# the plot function directly plot the graph using matplotlib

df1.plot()
```

```python
# To do interactive ploting using iplot in ploty 
df1.iplot(kind='line')
df1.iplot(title='same o/p as line plot') 
```

```python
#iplot Scatter ploting 
df1.iplot(kind='scatter',x='A',y='D',title='plot-1')

## iplot SCATTER PLOTING

df1.iplot(kind='scatter',x='A',y='D',mode='markers',title='plot-2',size=5)
```

```python
##iplot barploting
df1.iplot(kind='bar')
df2.iplot(kind='bar',x='Category',y='values')
df1.sum().iplot(kind='bar')
```

```python
#boxplot
df1.iplot(kind='box',title='data-1',
    xTitle='Cloumns',
    yTitle='values')
```

```python
#3d surface plot for data


df3=pd.DataFrame({
    'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]
})
df3.head()

df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
#3d surface plot for data
df1.iplot(kind='surface')

df1.iplot(kind='surface',x='A',y='B',colorscale='rdylbu')
```

```python
## hist
df1.iplot(kind='hist',bins=25)
df1['D'].iplot(kind='hist',bins=25)
```

```python
# spread plot

df1[['A','B']].iplot(kind='spread')
```

```python
import plotly.express as px
import pandas as pd


# Creating a spread plot using plotly
fig = px.line(df1, x=df1.index, y=['A', 'D'], title='Spread')
fig.show()
```

```python
## Bubble plot
df1.iplot(kind='bubble',x='A',y='B',size='D',color='cadetblue')
```
