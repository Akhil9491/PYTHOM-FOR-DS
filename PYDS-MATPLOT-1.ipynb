```python
x = np.arange(0,52,2)
y= x**2
y
```

```python
## function

plt.plot(x,y)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('x&y value')
plt.show()
```

```python
## sub ploting

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
##object oriented api-method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.5,0.8])
axes.plot(x,y)
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('label')
#axes.show()
```

```python
###  legends ##

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,'r',label = 'squres of x')
ax.plot(x,x**1.8,'b',label = 'root of 1.8 of x')

ax.legend(loc=0) 
```

```python
## marking legends location in another way
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color='red',linewidth=3,alpha=0.5,label = 'squres of x') ## alpha for line transparency starts form 0 to 1 value.

ax.plot(x,x**1.8,color='blue',linewidth=2,linestyle='-.',label = 'root 1.8 of x') ## linestyle/ls for styling the line

ax.plot(x,x**1.6,color='blue',linewidth=3,linestyle='--',alpha= 0.5,marker ='o',markersize= 4.5,markerfacecolor = 'red',label = 'root of 1.5 of x') 
## Marker to denote the points   ## alpha for line transperency  ## Markersize for size of the point
## LS for styling the line  ## linewidth for to increse or decrese the width of graph line  ##Markerfacecolor for to change the colour of the point color

ax.plot(x,x**1.4,color='black',linewidth=2,alpha=1,marker = '*',markerfacecolor='w',markersize=6,markeredgecolor='red',label = 'root 1.6 of x') 
## markeredgecolor for edge of point color 

ax.legend(loc=(0.0,0.8))
```
